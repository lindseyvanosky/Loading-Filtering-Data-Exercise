```python
import pandas as pd
filename = "/content/drive/MyDrive/Coding Dojo/02 Week 2: Pandas/Data Sets/RailsToTrails_National Count Data_week (2).xlsx"
bike_df = pd.read_excel(filename)
bike_df.head() 
```

```python
bike_df.dtypes
```

```python
#Select just the data after July 31, 2020
#I think Week 1 is supposed to be 2019-12-30 for the timing to make sense
autumn_filter = bike_df["Week of"] > "2020-07-31"
autumn_df = bike_df.loc[autumn_filter, :]
autumn_df
```

```python
#Select data where "Change 2019-2020" is greater than 100%
growth_filter = bike_df["Change 2019-2020"] >= 1.00
growth_df = bike_df.loc[growth_filter, :]
growth_df
```

```python
#Combine Filters!
autumn_best_df = bike_df.loc[growth_filter & autumn_filter, :]
autumn_best_df
#There were not any weeks in Autumn where change was > %100
```
